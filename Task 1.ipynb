```python
cat_cols = df.select_dtypes(include=['object']).columns
cols= pd.DataFrame({"count": df[cat_cols].isnull().sum(), "column": cat_cols})
```

```python
null_cols = list(cols[cols["count"] > 0].column)
```

```python
df_ohe = pd.get_dummies(df, columns=null_cols, dummy_na=True)
```

```python
non_null = df_ohe.select_dtypes(include=['object']).columns
```

```python
for col in non_null:
    print(df_ohe[col].value_counts())
```

We dont want to One Hot Encode binary variables. 

```python
df_ohe = pd.get_dummies(df_ohe, columns=['COD'], dummy_na=True)
```

```python
df_ohe.columns
```

```python
bin_cols = df_ohe.select_dtypes(include=['object']).columns
```

```python
df_ohe[bin_cols] = df_ohe[bin_cols].apply(lambda x: pd.Categorical(x).codes)
```

```python
df_ohe.to_csv('cleaned.csv')
```

```python
from sklearn.impute import KNNImputer
def impute_df(df_to_impute) -> pd.DataFrame:
    imputer = KNNImputer(n_neighbors=20, weights="distance")
    df_imputed = pd.DataFrame(imputer.fit_transform(df_to_impute),columns = df_to_impute.columns)
    return df_imputed
```

```python
df_ohe = impute_df(df_ohe)
```

```python
df_ohe.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numeric_cols = df_ohe.select_dtypes(include=[np.int64, np.float64]).columns
df_ohe[numeric_cols] = scaler.fit_transform(df_ohe[numeric_cols])
```

# Data Modeling 

```python
# import model and matrics
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, RepeatedStratifiedKFold, StratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix,roc_curve, roc_auc_score, precision_score, recall_score, precision_recall_curve
from sklearn.metrics import f1_score
# split dataset into x,y
x = df_ohe.drop(columns=['Class', 'id'],axis=1)
y = df_ohe['Class']
# train-test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=13)
```

```python
df_ohe['Class'].value_counts()/df.shape[0]
```

```python
# define class weights
w = {0:93, 1:7}
# define model
lg2 = LogisticRegression(random_state=13, class_weight=w, max_iter=2000)
# fit it
lg2.fit(X_train,y_train)
# test
y_pred = lg2.predict(X_test)
# performance
print(f'Accuracy Score: {accuracy_score(y_test,y_pred)}')
print(f'Confusion Matrix: \n{confusion_matrix(y_test, y_pred)}')
print(f'Area Under Curve: {roc_auc_score(y_test, y_pred)}')
print(f'Recall score: {recall_score(y_test,y_pred)}')
```
